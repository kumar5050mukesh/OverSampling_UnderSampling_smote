```python
#REGRESSION IMPUTATION
import pandas as pd
from sklearn.linear_model import LinearRegression

"""first we create a dataframe with three columns A,B,C"""
data = pd.DataFrame({
    'A': [1, 2, 3, None, 5,8],
    'B': [None, None, 8, 9, 10,34],
    'C': [11, 12, 13, 14, 15,25]
})

# create a regression model using the values in columns A and C to predict the missing values in column B
reg = LinearRegression()
train_data = data.dropna()

reg.fit(train_data[['A', 'C']], train_data['B'])


test_data = data[data['B'].isna()]
print(test_data)
test_data['B'] = reg.predict(test_data[['A', 'C']])
print(test_data["B"])


data_imputed = pd.concat([train_data, test_data])



```
